```python
import numpy as np
import re
import math

file = open("05.txt", "r")
lines = file.readlines()
file.close()

n = math.floor(len(lines[0]) / 4)
stack = np.full(n, "", dtype=object)

def printStack(stack):
  res = ""
  for s in range(len(stack)):
    res += stack[s][0]
  res
  print(stack, res)

i = 0
while i < len(lines):
  st = lines[i];
  items = [st[j:j+4] for j in range(0, len(st), 4)]
  if len(items) < n or items[0][1] == '1':
    break
  for j in range(n):
    if items[j][1] != " ":
      stack[j] += items[j][1]
  i += 1

startMoveIndex = i

def processLines(stack, shouldInverse):
  i = startMoveIndex
  while i < len(lines):
    # move 1 from 2 to 1
    parts = re.match("move (\d+) from (\d+) to (\d+)", lines[i])
    if parts:
      vals = list(map(int, parts.groups()))
      cnt = vals[0]
      f = vals[1] - 1
      to = vals[2] - 1
      if shouldInverse:
        stack[to] = stack[f][0:cnt][::-1] + stack[to]
      else:
        stack[to] = stack[f][0:cnt] + stack[to]
      stack[f] = stack[f][cnt:]
    i += 1
  return stack

print("Part1:")
printStack(processLines(stack.copy(), True))
print("Part2:")
printStack(processLines(stack.copy(), False))
```

```python
import torch
import numpy as np
import re
import math

file = open("09.txt", "r")
lines = file.readlines()
file.close()

n = 10
pos = torch.tensor(np.zeros(shape=(n,2)), dtype=torch.int32) 
delta = {
    "R": torch.tensor([1, 0]),
    "L": torch.tensor([-1, 0]),
    "U": torch.tensor([0, 1]),
    "D": torch.tensor([0, -1]),
}

def updatePos(hPos, tPos): 
  d = hPos - tPos
  if torch.any(d > 1) or torch.any(d < -1):
    grad = torch.maximum(torch.minimum(d, torch.tensor([1, 1])), torch.tensor([-1, -1]))
    return tPos + grad
  return tPos

dict = {'0-0': True}
for l in lines:
  vals = l.strip().split(" ")
  for i in range(int(vals[1])):
    pos[0] += delta[vals[0]]
    for j in range(1, n):
      pos[j] = updatePos(pos[j-1], pos[j])
    dict[
        str(pos[n-1][0].item()) + "-" + str(pos[n-1][1].item())
        ] = True
  
print('Part2', len(dict))
```

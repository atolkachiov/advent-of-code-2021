```python
import torch
import numpy as np
import re
import math

file = open("10.txt", "r")
lines = file.readlines()
file.close()

x = 1
part1 = 0
c = 0

crt = torch.tensor((), dtype=torch.float64).new_zeros(240)

def nextCycle():
  global c, part1, x, crt
  c += 1
  if c > 1 and (c - 20) % 40 == 0:
    part1 += c * x
  col = (c - 1) % 40
  crt[c - 1] = 1 if abs(col - x) <= 1 else 0
  # print(c, x)

for l in lines:
  vals = l.strip().split(" ")
  if vals[0] == 'addx':
    nextCycle()
    nextCycle()
    x += int(vals[1])
  else:
    nextCycle()

print('Part2')

part2 = torch.reshape(crt, (6,40))
for row in part2:
  print("".join(list(map(
      lambda x: '.' if x == 0 else '#',
      row
  ))))

```

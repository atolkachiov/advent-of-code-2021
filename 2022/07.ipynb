```python
import numpy as np
import re
import math

file = open("07.txt", "r")
lines = file.readlines()
file.close()

dsk = {"dirs":{}, "files": {}, "size":0}
cur = dsk

for i in range(len(lines)):
  st = lines[i].strip()
  if st[0:4] == '$ cd':
    where = st[5:]
    print('-- cd', where)
    if where == '..':
      cur = cur['root']
    elif where == '/':
      pass
    else:
      cur = cur['dirs'][where];
  if st == '$ ls':
    print('-- ls')
    i += 1
    while i < len(lines) and lines[i][0] != '$':
      s = lines[i].strip()
      if s[0:4] == 'dir ':
        dirname = s[4:]
        print("::dir", dirname)
        if not (dirname in cur['dirs']):
          cur['dirs'][dirname] = {"dirs":{}, "files": {}, "size":0, "root": cur}
      else:
        parts = s.split(' ')
        if len(parts):
          size = parts[0]
          filename = parts[1]
          print("::file", filename, size)
          if filename not in cur['files']:
            cur['files'][filename] = int(size)
      i += 1

part1Size = 0
allDirSizes = []

def walkSize(cur):
  global part1Size
  size = 0
  for file in cur['files']:
    size += cur['files'][file]
  for dir in cur['dirs']:
    size += walkSize(cur['dirs'][dir])
  allDirSizes.append(size)
  if size <= 100000:
    part1Size += size
  return size

totalSize = walkSize(dsk)

print('Part 1', part1Size)


toDelete = totalSize - 40000000
allDirSizes.sort()

print('allDirSizes', allDirSizes)
print('totalSize', totalSize)
print('toDelete', toDelete)

for curSize in allDirSizes:
  if curSize >= toDelete:
    print("Part2", curSize)
    break
```

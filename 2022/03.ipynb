```python
import numpy as np

file = open("03.txt", "r")
lines = file.readlines()
file.close()

def priority(ch):
  p = 0
  if ch >= ord('a') and ch <= ord('z'):
    p = ch - ord('a') + 1
  if ch >= ord('A') and ch <= ord('Z'):
    p = ch - ord('A') + 27
  return p

score = 0
for l in lines:
  st = l.strip();
  p1 = st[0:int(len(st)/2)]
  p2 = st[int(len(st)/2):]
  for ch in np.intersect1d(
    np.fromstring(p1, dtype=np.uint8),
    np.fromstring(p2, dtype=np.uint8)
  ):
    score = score + priority(ch)
score
```

```python
# Part 2

score = 0

for bucket in np.array_split(lines, int(len(lines) / 3)):
  intersect = np.intersect1d(
    np.intersect1d(
      np.fromstring(bucket[0].strip(), dtype=np.uint8),
      np.fromstring(bucket[1].strip(), dtype=np.uint8)
    ),
    np.fromstring(bucket[2].strip(), dtype=np.uint8)
  )
  score = score + priority(intersect[0])
score
```
